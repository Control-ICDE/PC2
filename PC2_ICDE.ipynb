```python
# Importamos la data de GidHub
import pandas as pd
datos = pd.read_excel('https://github.com/Control-ICDE/PC2/blob/main/HDI.xlsx?raw=true')
from tabulate import tabulate
print(datos)
```

```python
# Eliminamos las cuatro primeras lienas de la tabla porque no tienen informacion
datos_limpios = datos.iloc[3:]
print(datos_limpios)
```

```python
#Conservamos las dos primeras columnas que comienzan con NaN
colum_mantener = datos_limpios.columns[:2]
columnas_no_nan = datos_limpios.loc[:, datos_limpios.iloc[0].notna()].iloc[:, 2:]
datos_finales = pd.concat([datos_limpios[colum_mantener], columnas_no_nan], axis=1)
print(datos_finales)
```

```python
# Ahora, borramos las filas con puro NaN como valor a execcion de la segunda fila
datos_limpios = datos_finales[~(datos_finales.iloc[:, 2:].isna().all(axis=1) & datos_finales.iloc[:, 1].notna())]
print(datos_limpios)
```

```python
# A continuacion, eliminaremos las filas que tengan puro NaN
datos_limpios = datos_limpios.dropna(how='all')
print(datos_limpios)
```

```python
#Ahora, dejamos en blanco todas las celdas con NaN y tambien los que tengan ".." y "-"
datos_limpios = datos_limpios.fillna('')
datos_limpios = datos_limpios.replace({'..': '', '—': ''})
datos_limpios
```
